## 2) Handling the pdf files

```
#Checking the content in each page
page = pdf.getPage(1)
print(page.extractText())
print(page.extractText().replace('_', ''))
print(re.split(r'[\s\n]\s*\n*',page.extractText().replace('_', '')))
```

```
ticket = []
date = []
location = []
des = []
cnt = []
cpm = []
mis = []

for i in range(pdf.numPages):
  cnt.append(i)
for i in range(pdf.numPages):
  cpm.append('')
for i in range(pdf.numPages):
  des.append('')

for j in range(pdf.numPages):
  page = pdf.getPage(j)
  m = re.split(r'[\s\n]\s*\n*',page.extractText().replace('_', ''))
  
  while(m.count(' ')):
    m.remove(' ')
  while(m.count('')):
    m.remove('')
  
  i = 0
  n = []
  if m[i][0] == "T":
    i += 1
    if m[i] != '#':
      i += 1
    while(m[i] != "Date:" and m[i] + m[i+1] != "Date:"):
      n.append(m[i])
      i += 1
    n = " ".join(n)
    ticket.append(n)
  
  n = []
  if m[i] == "Date:" or m[i] + m[i+1] == "Date:":
    i += 1
    if m[i-1]+m[i] == "Date:":
      i += 1
    while(m[i] != 'City/State/Zip:' and m[i]+m[i+1] != 'City/State/Zip:'):
      n.append(m[i])
      i += 1
    n = " ".join(n)
    date.append(n)
  
  n = []
  if m[i] == 'City/State/Zip:' or m[i]+m[i+1] == 'City/State/Zip:':
    i += 1
    if m[i-1]+m[i] == 'City/State/Zip:':
      i += 1
    while(m[i] != 'Company' and m[i] != 'Description' and m[i] + m[i+1] != 'Description'):
      n.append(m[i])
      i += 1
      if i+1 == len(m):
        break
    n = " ".join(n)
    location.append(n)
  if i+1 == len(m):
    continue
  
  n = [] 
  if m[i] == 'Company':
    while(m[i][len(m[i]) - 1] != ":"):
      i += 1
    while(m[i] != 'Description' and m[i] + m[i+1] != 'Description'):
      if m[i][len(m[i]) - 1] == ":":
        i += 1
      n.append(m[i])
      i += 1
    n = " ".join(n)
    cpm[j] = n
  if i+1 == len(m):
    continue
  
  n = []
  if m[i] == "Description" or m[i] + m[i+1] == "Description":
    i += 1
    if m[i-1]+m[i] == 'Description':
      i += 1
    while(i <= len(m)-1):
      n.append(m[i])
      i += 1
    n = " ".join(n)
    des[j] = n

print(len(ticket))
print(len(date))
print(len(location))
print(len(cpm))
print(len(des))
```
